```python
from gurobipy import *


MP=Model("benders_MP")

""" create decision variables """
y = MP.addVar(lb=0,ub=1000,vtype= GRB.INTEGER,name='y')
q = MP.addVar(lb=0,ub=GRB.INFINITY,vtype=GRB.INTEGER,name='q')
z = MP.addVar(lb=0,ub=GRB.INFINITY,vtype=GRB.INTEGER, name='z')
MP.setObjective(z, GRB.MAXIMIZE)

""" add constraints """
MP.addConstr(z==q+1.045*y,name='constraint_1')
MP.addConstr(z<=1055+1.045*y,name='optimality cut 1')
MP.addConstr(z<=1100-0.055*y,name='optimality cut 2')
MP.addConstr(z<=1055+0.044*y,name='optimality cut 3')
MP.addConstr(z<=1065-0.005*y,name='optimality cut 4')
MP.addConstr(z<=1058+0.015*y,name='optimality cut 5')
MP.addConstr(z<=1061+0.005*y,name='optimality cut 6')

MP.optimize()
print('\n\n\n')
print('Obj:', MP.ObjVal)
print('z = %4.1f' % (z.x))
print('y = %4.1f' % (y.x))
print('q = %4.1f' % (q.x))
```

```python
from gurobipy import *

Dual_SP=Model("benders_Dual_SP")
y_bar=400
""" create decision variables """
alpha_0 = Dual_SP.addVar(lb=0,ub=GRB.INFINITY,vtype=GRB.CONTINUOUS,name='alpha_0')
alpha={}
for i in range(1,11):
    alpha[i] = Dual_SP.addVar(lb=0, ub=GRB.INFINITY, vtype=GRB.CONTINUOUS, name='alpha_' + str(i))

""" create objective function """
obj = LinExpr()
obj.addTerms(1000-y_bar, alpha_0)
for i in range(1,11):
    obj.addTerms(100, alpha[i])
Dual_SP.setObjective(obj, GRB.MINIMIZE)

""" add constraints 1-10 """
for i in range(1,11):
    Dual_SP.addConstr(alpha_0+alpha[i]>=1+0.01*(i), name='constraint_'+str(i))

Dual_SP.optimize()
print('\n\n\n')
print('Model status:', Dual_SP.status)
if(Dual_SP.status == 2):
    print('Obj:', Dual_SP.ObjVal)
    print('{} = {}'.format(alpha_0.varName, alpha_0.x))
    for i in range(1,11):
        if(alpha[i].x > 0):
            print('{} = {}'.format(alpha[i].varName, alpha[i].x))
```

# 运筹优化算例

```python
from gurobipy import *

model = Model("benders-decomposition")

""" create decision variables """
y={}
for i in range(1,6):
    y[i] = model.addVar(lb=0, ub=1, vtype=GRB.BINARY, name='y_' + str(i))
x={}
for i in range(1,6):
    x[i] = model.addVar(lb=0, ub=GRB.INFINITY, vtype=GRB.CONTINUOUS, name='x_' + str(i))

""" set objective function """
obj = LinExpr()
for i in range(1,6):
    obj.addTerms(7, y[i])
for i in range(1,6):
    obj.addTerms(1, x[i])
model.setObjective(obj, GRB.MINIMIZE)

""" add constraints """
model.addConstr(x[1]+x[4]+x[5]==8,name='constraint_1')
model.addConstr(x[2]+x[5]==3,name='constraint_2')
model.addConstr(x[3]+x[4]==5,name='constraint_3')
model.addConstr(x[1]<=8*y[1],name='constraint_4')
model.addConstr(x[1]<=8*y[1],name='constraint_5')
model.addConstr(x[2]<=3*y[2],name='constraint_6')
model.addConstr(x[3]<=5*y[3],name='constraint_7')
model.addConstr(x[4]<=5*y[4],name='constraint_8')
model.addConstr(x[5]<=5*y[5],name='constraint_9')

model.optimize()
print('\n\n\n')
print('Obj:', model.ObjVal)
for i in range(1,6):
    print('Fund ID {}: amount: {}'.format(i,y[i].x))
for i in range(1,6):
        print('Fund ID {}: amount: {}'.format(i,x[i].x))
```

```python
from gurobipy import *

MP=Model("benders_MP")

""" create decision variables """
y={}
for i in range(1,6):
    y[i]=MP.addVar(lb=0,ub=1,vtype= GRB.BINARY,name='y_'+str(i))
q = MP.addVar(lb=0,ub=GRB.INFINITY,vtype=GRB.CONTINUOUS,name='q')
z = MP.addVar(lb=0,ub=GRB.INFINITY,vtype=GRB.CONTINUOUS, name='z')

MP.setObjective(z, GRB.MINIMIZE)

""" add constraints """
MP.addConstr(z==q+7*y[1]+7*y[2]+7*y[3]+7*y[4]+7*y[5],name='constraint_1')

MP.optimize()
print('\n\n\n')
print('Obj:', MP.ObjVal)
for i in range(1,6):
    print('Fund ID {}: y{}={}'.format(i,i,y[i].x))
print('q = %4.1f' % (q.x))
print('z = %4.1f' % (z.x))
```

```python
from gurobipy import *
y1_bar=0
y2_bar=0
y3_bar=0
y4_bar=0
y5_bar=0
Dual_SP=Model("benders_Dual_SP")

""" create decision variables """
alpha = {}
beta = {}
for i in range(1,4):
    alpha[i] = Dual_SP.addVar(lb=-GRB.INFINITY, ub=GRB.INFINITY, vtype=GRB.CONTINUOUS, name='alpha_' + str(i))
for i in range(1,6):
    beta[i] = Dual_SP.addVar(lb=-GRB.INFINITY, ub=0, vtype=GRB.CONTINUOUS, name='beta_' + str(i))

""" create objective function """
obj = LinExpr()
obj.addTerms(8, alpha[1])
obj.addTerms(3, alpha[2])
obj.addTerms(5, alpha[3])
obj.addTerms(8*y1_bar, beta[1])
obj.addTerms(3*y2_bar, beta[2])
obj.addTerms(5*y3_bar, beta[3])
obj.addTerms(5*y4_bar, beta[4])
obj.addTerms(5*y5_bar, beta[5])

Dual_SP.setObjective(obj, GRB.MAXIMIZE)

""" add constraints 1-5 """
Dual_SP.addConstr(alpha[1]+beta[1]<=1,name='constraint_1')
Dual_SP.addConstr(alpha[2]+beta[2]<=1,name='constraint_2')
Dual_SP.addConstr(alpha[3]+beta[3]<=1,name='constraint_3')
Dual_SP.addConstr(alpha[1]+alpha[3]+beta[4]<=1,name='constraint_4')
Dual_SP.addConstr(alpha[1]+alpha[2]+beta[5]<=1,name='constraint_5')

Dual_SP.optimize()
print('\n\n\n')
print('Model status:', Dual_SP.status)
if(Dual_SP.status == 2):
    print('Obj:', Dual_SP.ObjVal)
    for i in range(1,4):
            print('{} = {}'.format(alpha[i].varName, alpha[i].x))
    for i in range(1,6):
            print('{} = {}'.format(beta[i].varName, beta[i].x))
```

```python
Dual_SP.setParam('InfUnbdInfo', 1)
Dual_SP.optimize()

print('\n\n\n')
print('Model status:', Dual_SP.status)
if (Dual_SP.status == 2):
    print('Obj:', Dual_SP.ObjVal)
    for i in range(1, 4):
        print('{} = {}'.format(alpha[i].varName, alpha[i].x))
    for i in range(1, 6):
        print('{} = {}'.format(beta[i].varName, beta[i].x))
else:
    print('The model is infeasible or unbounded.')
    for i in range(1, 4):
        print('{} = {}'.format(alpha[i].varName, alpha[i].UnbdRay))
    for i in range(1, 6):
        print('{} = {}'.format(beta[i].varName, beta[i].UnbdRay))#注意
```
